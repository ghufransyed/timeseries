```python
class TSAE(torch.nn.Module):
    def __init__(self,seqLen,latentDim=24):
        super().__init__()
        #latentDim = max(seqLen // 8,8)
        self.rnn = HeroRNN()
        self.mean = torch.nn.Linear(100,latentDim)
        self.logvar = torch.nn.Linear(100,latentDim)
        

        layers = []
        for a,b in [(latentDim,100),(100,200),(200,300)]:
            layers += bn_drop_lin(a,b,actn=torch.nn.ReLU())
        self.lin = torch.nn.Sequential(*layers)
        self.out = torch.nn.Linear(300,seqLen)

    def forward(self,ts):
        seqLen = ts.shape[1]
        ts, _ = self.rnn(ts.unsqueeze(-1))
        mean, logvar = self.mean(ts), self.logvar(ts)
        #mean = self.mean(ts)
        
        ls = mean
        if self.training:
            std = torch.exp(0.5 * logvar)
            eps = torch.randn_like(std)
            ls = eps.mul(std).add_(mean)
        #return self.out(self.lin(ls))          
        return self.out(self.lin(ls)), mean, logvar
```

```python
class VAELoss(torch.nn.Module):
    def __init__(self,beta=1.0):
        super().__init__()
        self.beta = beta
    def forward(self,p,target):
        pred,mean,logvar = p
        self.mse = torch.nn.functional.mse_loss(pred,target,reduction="sum")
        self.kld = self.beta * -0.5 * torch.sum(1+logvar-mean.pow(2)-logvar.exp())
        return self.mse + self.kld
```

```python
class PrintLosses(LearnerCallback):
    _order=-20
    def __init__(self,learn):
        super().__init__(learn)
    def on_train_begin(self, **kwargs):
        self.learn.recorder.add_metric_names(["mse","kld"])
    def on_epoch_end(self, **kwargs):
        self.learn.recorder.add_metrics([self.learn.loss_func.mse,self.learn.loss_func.kld])
```

```python
learnAE = Learner(dataAE,TSAE(len(data.train_ds[0][0].data)),loss_func=VAELoss(0.25),callback_fns=PrintLosses)
#learnAE = Learner(dataAE,TSAE(len(data.train_ds[0][0].data)),loss_func=F.mse_loss)
```

```python
def create_head_1d(nf:int, nc:int, lin_ftrs:Optional[Collection[int]]=None, ps:Floats=0.5, bn_final:bool=False):
    "Model head that takes `nf` features, runs through `lin_ftrs`, and about `nc` classes."
    lin_ftrs = [nf, 512, nc] if lin_ftrs is None else [nf] + lin_ftrs + [nc]
    ps = listify(ps)
    if len(ps)==1: ps = [ps[0]/2] * (len(lin_ftrs)-2) + ps
    actns = [nn.ReLU(inplace=True)] * (len(lin_ftrs)-2) + [None]
    layers = []
    for ni,no,p,actn in zip(lin_ftrs[:-1],lin_ftrs[1:],ps,actns):
        layers += bn_drop_lin(ni,no,True,p,actn)
    if bn_final: layers.append(nn.BatchNorm1d(lin_ftrs[-1], momentum=0.01))
    return nn.Sequential(*layers)
```

```python
class HeroRNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.rnn1 = torch.nn.LSTM(1,100,1,batch_first=True)
        self.rnn2 = torch.nn.LSTM(100,100,1,batch_first=True)
        self.rnn3 = torch.nn.LSTM(100,100,1,batch_first=True)
        
    def forward(self, x):
        actvns = []
        
        output,(h_n,c_n) = self.rnn1(x)
        actvns.append(output)
        
        output,(h_n,c_n) = self.rnn2(output)
        actvns.append(output)
        
        output,(h_n,c_n) = self.rnn3(output)
        actvns.append(output)
        
        return h_n.squeeze(0), actvns
        
```

```python
class SidekickRNN(nn.Module):
    def __init__(self, numClasses):
        super().__init__()
        
        self.hero = HeroRNN()
        self.rnn1 = torch.nn.LSTM(1,100,1,batch_first=True)
        self.rnn2 = torch.nn.LSTM(200,100,1,batch_first=True)
        self.rnn3 = torch.nn.LSTM(200,100,1,batch_first=True)
        
        self.out = create_head_1d(200,numClasses)
                
    def forward(self,ts):
        ts = ts.unsqueeze(-1)
        pt, actvns = self.hero(ts)
        
        output, (h_n,c_n) = self.rnn1(ts)
        output, (h_n,c_n) = self.rnn2(torch.cat([output,actvns[0]],dim=2))
        output, (h_n,c_n) = self.rnn3(torch.cat([output,actvns[1]],dim=2))
        
        x = torch.cat([pt,h_n.squeeze(0)],dim=1)
        return self.out(x)
```
