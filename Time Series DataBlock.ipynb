## Time Series data_block extension
The following two classes extend the fastai data_block api to handle time series. This makes it simple to create databunches for any task. The `TimeSeriesList` class includes a `label_from_self` option for autoencoders.

The docs for the api are here:
<https://docs.fast.ai/data_block.html>

I followed this tutorial to implement the custom classes:
<https://docs.fast.ai/tutorial.itemlist.html>

```python
class TimeSeriesItem(ItemBase):
    def __init__(self,ts):
        self.data = self.obj = torch.tensor(ts,dtype=torch.float)
        
    def show(self, ax, title="", **kwargs):
        ax.plot(self.data)
        ax.set_title(title)
        
    def __str__(self):
        return f"Time series of size {self.data.shape}"
```

```python
class TimeSeriesList(ItemList):
    def __init__(self,items,labeled=-1,**kwargs):
        self.labeled = labeled
        super().__init__(items, **kwargs)
    
    def new(self, items, **kwargs):
        return super().new(items,labeled=self.labeled,**kwargs)
    
    @classmethod
    def from_csv_list(cls,csv_paths,labelCol=-1,header=None,sep="\t",**kwargs):
        dfs = []
        nl = []
        for f in csv_paths:
            dfs.append(pd.read_csv(f,header=header,sep=sep))
            nl += [f.name]*len(dfs[-1])
        df = pd.concat(dfs)
        return cls(df.values,labeled=labelCol,xtra=np.array(nl))        
    
    @classmethod
    def from_numpy(cls,nparray,**kwargs):
        return cls(items=nparray,**kwargs)
    
    def get(self, i):
        a = super().get(i).astype(np.float)
        return TimeSeriesItem(np.delete(a,self.labeled,0) if self.labeled >= 0 else a)
    
    def split_by_csv_name(self, valid_names):
        return self.split_by_idx(np.where(np.isin(self.xtra,valid_names))[0])
    
    def label_from_col(self, **kwargs):
        return self.label_from_func(lambda o: o[self.labeled],label_cls=CategoryList)
    
    def label_from_self(self, **kwargs):
        return self._label_list(x=self,y=self)
        
    def reconstruct(self, t):
        return TimeSeriesItem(t.numpy())
    
    def show_xys(self, xs, ys, figsize=(12,10),plotY=True, **kwargs):
        rows = int(math.sqrt(len(xs)))
        fig, axs = plt.subplots(rows,rows,figsize=figsize)
        for i, ax in enumerate(axs.flatten() if rows > 1 else [axs]):
            xs[i].show(ax=ax)
            if plotY and isinstance(ys[0],TimeSeriesItem):
                ys[i].show(ax=ax)
                
    def show_xyzs(self, xs, ys, zs, attn=None,figsize=(12,10), **kwargs):
        rows = int(math.sqrt(len(xs)))
        fig, axs = plt.subplots(rows,rows,figsize=figsize)
        for i, ax in enumerate(axs.flatten() if rows > 1 else [axs]):
            xs[i].show(ax=ax)
            if isinstance(zs[0],TimeSeriesItem):
                zs[i].show(ax=ax)
            if attn:
                ax.plot(10*attn.attnSave[i].flatten())
```

```python
# Simple utility class to manage the datasets in the UCR archive, datasets available for download here: 
# https://www.cs.ucr.edu/~eamonn/time_series_data_2018/

class UCRArchive():
    def __init__(self,path=Path("UCRArchive_2018")):
        self.path = path if isinstance(path,Path) else Path(path)
        self.datasets = [f for f in self.path.iterdir() if f.is_dir()]
        self.dsdict = {f.name : f for f in self.datasets}
        
    def get(self,name):
        f = self.dsdict[name]
        return f
    
    def get_csv_files(self,name):
        f = self.dsdict[name]
        return list(f.glob("*.tsv"))
    
    def list_datasets(self):
        return [d.name for d in self.datasets]
    
    def category_distribution(self,name):
        path = self.dsdict[name]
        trainDF,testDF = [pd.read_csv(p,header=None,sep="\t") for p in [path/f"{name}_TRAIN.tsv",path/f"{name}_TEST.tsv"]]
        trainDF.iloc[:,0].hist()
        testDF.iloc[:,0].hist()
```

Now we have access to the full fastai machinery. First, create a databunch for a classifier.

```python
data = (TimeSeriesList.from_csv_list(ucr.get_csv_files("ElectricDevices"),labelCol=0)
                      .split_by_csv_name(["ElectricDevices_TEST.tsv"])
                      .label_from_col() #The first column in the tsv files contains the class label
                      .databunch(num_workers=0)) #bug on windows if num_workers is higher than 0
```

Because the `ItemBase` subclass includes a show method, we can use the databunch/learner show methods.

```python
data.show_batch()
```

```python
#Pytorch model
class TSClassifier(torch.nn.Module):
    def __init__(self,numClasses):
        super().__init__()
        self.rnn = torch.nn.LSTM(1,200,2,batch_first=True,dropout=0.5)
        self.attn = torch.nn.Linear(200,1,bias=False)
        self.out = torch.nn.Linear(200,numClasses)
        
    def forward(self, ts):
        output, _ = self.rnn(ts.unsqueeze(-1))
        a = torch.softmax(self.attn(output),dim=1)
        self.attnSave = a
        output = output * a
        output = output.sum(dim=1)
        return self.out(output)
```

```python
#Create a fastai learner
learn = Learner(data,TSClassifier(data.train_ds.c),loss_func=F.cross_entropy,metrics=[accuracy])
```

```python
# Run one batch through the model to make sure the output size is correct
learn.model(next(iter(data.train_dl))[0]).size()
```

```python
#Learning rate finder
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10,1e-2,wd=0.1)
```

```python
#Plot input in blue and attention in orange
learn.show_results(DatasetType.Train,attn=learn.model,rows=9)
```
